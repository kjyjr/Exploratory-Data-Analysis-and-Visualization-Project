```python
# Dependencies
#%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# National Average Annual Emissions: All EV - 2817 pounds CO2. Plug in hybrid 4824. Hybrid 6898. Gasoline 12594.
# Create a dataframe to show X-axis as type of vehicle, y-axis annual CO2 emissions.
```

```python
CO2_emissions = [2817, 4824, 6898, 12594]
x_axis = np.arange(len(CO2_emissions))

plt.bar(x_axis, CO2_emissions, color='g', alpha=0.5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Electric (EV)", "Plug-In Hybrid Electric (PHEV) ", "Hybrid Electric (HEV)", "Gasoline"], rotation="vertical")
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(CO2_emissions)+1000)
plt.title("Average Annual CO2 Emissions")
plt.xlabel("Vehicle Type")
plt.ylabel("Pounds of CO2 Equivalent")

plt.show()
```
