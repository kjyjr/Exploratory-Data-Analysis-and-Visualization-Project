```python
import pandas as pd
import matplotlib.pyplot as plt

#Reading CSV
gas_df = pd.read_csv("gas_prices.csv")
electricity_df = pd.read_csv("electricity_prices.csv")
```

```python

#figure and axis
plt.figure(figsize=(12, 6))
ax = plt.gca()

# Plot gas
plt.plot(gas_df['Date'], gas_df['Price'], label='Gas Prices', marker='o')

# Plot electricity
plt.plot(electricity_df['Date'], electricity_df['Price'], label='Electricity Prices', marker='x')

# labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Gas vs. Electricity Prices Over the Years')

# Customize the x-axis to show years
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Show the chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read gas prices data from CSV
gas_df = pd.read_csv("gas_prices.csv")

# Read electricity prices data from CSV
electricity_df = pd.read_csv("electricity_prices.csv")

# Create a figure and axis
plt.figure(figsize=(12, 6))
ax = plt.gca()

# Group and aggregate data by year and calculate the average price
gas_grouped = gas_df.groupby('Date')['Price'].mean()
electricity_grouped = electricity_df.groupby('Date')['Price'].mean() * 33.7  # Multiply electricity prices

# Plot gas prices
plt.plot(gas_grouped.index, gas_grouped, label='Gas Prices', marker='o')

# Plot electricity prices
plt.plot(electricity_grouped.index, electricity_grouped, label='Electricity Prices (multiplied by 33.7)', marker='x')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Gas vs. Electricity Prices Over Multiple Years')

# Customize the x-axis to show years
plt.xticks(gas_grouped.index)

# Add a legend
plt.legend()

# Show the chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Read the electricity prices data 
electricity_data = pd.read_csv("electric_state.csv")


```

```python
selected_states = ["California", "New York", "Texas", "Colorado", "Ohio"]
selected_data = electricity_data[electricity_data["State"].isin(selected_states)]
```

```python
import matplotlib.pyplot as plt

# Create the comparison chart
plt.figure(figsize=(10, 6))
plt.bar(selected_data["State"], selected_data["Price"])

# Customize the chart
plt.xlabel('State')
plt.ylabel('Electricity Price')
plt.title('Electricity Prices for Selected States')
plt.xticks(rotation=45)

# Show the chart
plt.tight_layout()
plt.show()
```
