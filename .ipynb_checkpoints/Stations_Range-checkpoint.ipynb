```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python

ev_range_path = "Sources/EVs with longest driving range.csv"
ev_range = pd.read_csv(ev_range_path)

ev_stations_path = "Sources/ev_fuel_stations_5.18.2023.csv"
ev_stations = pd.read_csv(ev_stations_path, low_memory=False)

tesla_stations_path = "Sources/tesla_stations_and_stores-2022-09-05.csv"
tesla_stations = pd.read_csv(tesla_stations_path, low_memory=False)
```

```python
vehicle_names = ev_range['CAR']
#x_axis = np.arange(len(vehicle_names))
x_axis = ev_range['CAR']
y_axis = ev_range['EPA ESTIMATED DRIVING RANGE (in miles)']

tick_locations = [value for value in x_axis]
plt.figure(figsize=(14, 4))
plt.bar(x_axis, y_axis, color='r', alpha=0.7, align='center')
plt.xticks(tick_locations, vehicle_names, rotation='vertical')
plt.xlim(-0.75, len(x_axis))

plt.title('ELECTRIC VEHICLE DRIVING RANGES')
plt.xlabel('Vehicle')
plt.ylabel('EPA ESTIMATED DRIVING RANGE (in miles)')

#plt.savefig('Images/ev_driving_ranges.png')
plt.show()
```

```python
driving_range = ev_range['EPA ESTIMATED DRIVING RANGE (in miles)']
x_axis = driving_range 
y_axis = ev_range['CAR']

tick_locations = np.arange(min(driving_range), max(driving_range)+1, 100)

plt.figure(figsize=(15, 10))
plt.barh(y_axis, x_axis, color='g', alpha=0.7, align='center')
plt.xticks(tick_locations, rotation='horizontal')
plt.ylim(-1, len(y_axis))

plt.title('ELECTRIC VEHICLE DRIVING RANGES')
plt.xlabel('EPA ESTIMATED DRIVING RANGE (in miles)')
plt.ylabel('Vehicle')
plt.gca().invert_yaxis()

for index, value in enumerate(x_axis):
    plt.text(value, index,
             str(value))
    
plt.savefig('Images/ev_driving_ranges.png')
plt.show()

#https://www.geeksforgeeks.org/draw-a-horizontal-bar-chart-with-matplotlib/
#https://www.statology.org/matplotlib-set-axis-ticks/
#https://www.bing.com/search?pglt=43&q=matplotlib+how+reverse+order+on+y+axis&cvid=4514088e44cb46f080f48da692b7cd01&gs_lcrp=EgZjaHJvbWUyBggAEEUYOdIBCTEyMzgxajBqMagCALACAA&FORM=ANNTA1&PC=HCTS
#https://www.geeksforgeeks.org/how-to-display-the-value-of-each-bar-in-a-bar-chart-using-matplotlib/?ref=lbp
```
