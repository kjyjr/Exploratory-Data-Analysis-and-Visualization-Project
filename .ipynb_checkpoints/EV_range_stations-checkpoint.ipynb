```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python

ev_range_path = "Sources/EVs with longest driving range.csv"
ev_range = pd.read_csv(ev_range_path)

ev_stations_path = "Sources/ev_fuel_stations_5.18.2023.csv"
ev_stations = pd.read_csv(ev_stations_path, low_memory=False)

tesla_stations_path = "Sources/tesla_stations_and_stores-2022-09-05.csv"
tesla_stations = pd.read_csv(tesla_stations_path, low_memory=False)
```

```python
ev_range_df = ev_range
ev_range
```

```python
ev_range_top20_df = ev_range.iloc[0:20, 0:4]
ev_range_top20_df
```

```python
driving_range = ev_range_top20_df['EPA ESTIMATED DRIVING RANGE (in miles)']
price_range = ev_range_top20_df['PRICE (in dollars)']

x1_axis = driving_range 
x2_axis = price_range
y_axis = ev_range_top20_df['CAR']

plt.figure(figsize=(15, 10))
plt.barh(y_axis, x1_axis, color='g', alpha=0.7, align='center')

tick_locations = [0, 150, 300, 450, 600]

plt.xticks(tick_locations, rotation='horizontal')
plt.ylim(-1, len(y_axis))

plt.title('Top 20 ELECTRIC VEHICLE DRIVING RANGES')
plt.xlabel('EPA ESTIMATED DRIVING RANGE (in miles)')
plt.ylabel('Vehicle')
plt.gca().invert_yaxis()
    
for index, value in enumerate(x1_axis):
    plt.text(value, index, str(value))

plt.axvline(x=300, color='m', linestyle='dotted')

plt.savefig('Images/ev_driving_ranges_top20.png')
plt.show()
```

```python
revised_ev_stations_df = ev_stations[['Station Name', 'Street Address', 'City', 'State', 'ZIP', 'Groups With Access Code', 
                                           'Access Days Time', 'EV Level2 EVSE Num', 'EV DC Fast Count', 'EV Other Info', 
                                           'EV Network', 'EV Network Web', 'Geocode Status', 'Latitude', 'Longitude',
                                          'EV Connector Types', 'Facility Type', 'EV Pricing']]
revised_ev_stations_df.head()
```

```python
ev_stations_state_counts_df = revised_ev_stations_df['State'].value_counts().to_frame().reset_index()
ev_stations_state_counts_df
```

```python
ev_stations_state_counts_top20_df = ev_stations_state_counts_df.iloc[0:20, 0:3]
ev_stations_state_counts_top20_df
```

```python
revised_ev_stations_df['City'].value_counts()
```

```python
x_axis = ev_stations_state_counts_top20_df['State']
y_axis = ev_stations_state_counts_top20_df['count']

state_names = ev_stations_state_counts_top20_df['State']
tick_locations = [value for value in x_axis]
plt.figure(figsize=(15, 7))

plt.bar(x_axis, y_axis, color='b', alpha=0.7, align='center')
plt.xticks(tick_locations, state_names, rotation='vertical')
plt.xlim(-0.75, len(x_axis))

for index, value in enumerate(y_axis):
    plt.text(index - 0.35, value + 5.0, str(value))
    
plt.title('TOP 20 STATES BY STATION TOTALS')
plt.xlabel('State')
plt.ylabel('Number of Stations')
    
plt.savefig('Images/ev_driving_ranges.png')
plt.show()

#Source for code in this cell:
#https://stackoverflow.com/questions/53066633/how-to-show-values-on-top-of-bar-plot
```

```python
tesla_stations_state_sort_clean_df = tesla_stations_state_sort_clean.value_counts().to_frame().reset_index()
tesla_stations_state_sort_clean_df
```

```python
tesla_stations_state_sort_clean_df = tesla_stations_state_sort_clean_df.rename(columns={'state': 'State', 'count': 'Tesla Count'})
tesla_stations_state_sort_clean_df
```

```python
ev_stations_state_counts_top20_df = ev_stations_state_counts_top20_df.rename(columns={'count': 'State Total Count'})
ev_stations_state_counts_top20_df
```

```python
state_tesla_counts_top20_df = pd.merge(ev_stations_state_counts_top20_df, tesla_stations_state_sort_clean_df, on='State', how='left')
state_tesla_counts_top20_df
```

```python
colors = ['blue', 'red']
state_tesla_counts_top20_plot = state_tesla_counts_top20_df.plot(kind='bar', color=colors, figsize=(10, 5))
state_tesla_counts_top20_plot.set_xticklabels(state_tesla_counts_top20_df['State'], rotation=0)   

plt.title('TOP 20 STATES BY CHARGING STATION TOTALS')
plt.xlabel('State')
plt.ylabel('Number of Stations')


plt.axhline(y=2000, color='c', linestyle='-')
#Source:
#https://stackoverflow.com/questions/33382619/plot-a-horizontal-line-on-a-given-plot

plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

colors = ['blue', 'red']
state_tesla_counts_top20_plot = state_tesla_counts_top20_df.plot(kind='bar', color=colors, figsize=(10, 5))
state_tesla_counts_top20_plot.set_xticklabels(state_tesla_counts_top20_df['State'], rotation=0)   

plt.title('TOP 20 STATES BY CHARGING STATION TOTALS', x=0.45, y=0.9)
plt.xlabel('State')
plt.ylabel('Number of Stations')
plt.axhline(y=2000, color='c', linestyle='-')
#Source:
#https://stackoverflow.com/questions/33382619/plot-a-horizontal-line-on-a-given-plot

plt.savefig('Images/state_tesla_counts_plot.png')

fig, (ax1, ax2) = plt.subplots(2)

x1 = ev_stations_cities_top20_df.iloc[:,0]
y1 = ev_stations_cities_top20_df.iloc[:,1]

ax1.bar(x1, y1, color='b')

ax1.set_title('TOP 20 CITIES BY CHARGING STATION TOTALS - US', x=0.55, y=0.8, fontsize=8, fontweight='bold')
#Source:
#https://www.geeksforgeeks.org/how-to-adjust-title-position-in-matplotlib/

ax1.set_xticklabels(x1, rotation=45, ha='right')
ax1.set_ylabel('# of Stations')
ax1.axhline(y=500, color='red')
#Source:
#https://www.bing.com/search?q=how+set+a+horizontal+line+in+subplots+matplotlib&qs=NWU&pq=how+set+a+horizontal+line+in+subplot&sk=NW4NWU1&sc=9-36&cvid=3349695708B94E2CA90A7ED0EAEBCB54&FORM=QBRE&sp=6&ghc=1&lq=0

x2 = tesla_stations_us_cities_top20_df.iloc[:,0]
y2 = tesla_stations_us_cities_top20_df.iloc[:,1]

ax2.bar(x2, y2, color='r')
ax2.set_title('TOP 20 CITIES BY CHARGING STATION TOTALS - TESLA', x=0.55, y=0.8, fontsize=8, fontweight='bold')
ax2.set_xticklabels(x2, rotation=45, ha='right')
ax2.set_yticks([0, 50, 100, 150, 200])
ax2.axhline(y=25, color='blue')

plt.savefig('Images/us_tesla_counts_plot.png')

plt.tight_layout()
plt.show()
```
